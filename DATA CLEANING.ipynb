# DATA CLEANING

For this week I will be cleaning up the data that I received from various souses. The good this is that majority of the data for schools are very clean to begin with. This means that my data cleaning can be kept to a minimum.

The data sets I will be using are as follows:

    IRS's Income by zipcode 
    Graduation rates by school
    School Zip Code
    population
    city zip

The school's Zip Code file did not come in a CSV and to import it to a CSV I took to excel and created a CSV version of the excel file.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")


# HERE I IMPORTED DATA FROM NYSTATE DATA FOR
# GRADUATION RATES BY SCHOOL (NEED TO FIND A WAY TO ASSIGN ZIP CODE TO THEM)
grad2016 = pd.read_csv('SCHOOL_DATA/GRAD_RATE_AND_OUTCOMES_2016.csv', error_bad_lines = False)
grad2015 = pd.read_csv('SCHOOL_DATA/GRAD_RATE_AND_OUTCOMES_2015.csv' , error_bad_lines = False)

schoolZip = pd.read_csv('GEO_DATA/schoolDistZip.csv', error_bad_lines = False)

population = pd.read_csv('POPULATION_DATA/pop-by-zip-code.csv')
cityZip = pd.read_csv('GEO_DATA/cityZip.csv')


# The data contains different aggregated values from statewide, district, individual school, 
# county, etc... This means a lot of information is present within the data set. 
# The only thing is I only require 1 aggregation type and that is the district aggregation. 
# So below I took it upon myself to retrieve the aggregation type "District" and "SCHOOL" 
# from the dataframe.

grad2015Schools = grad2015.loc[grad2015['AGGREGATION_TYPE'] == 'School']
grad2016Schools = grad2016.loc[grad2016['AGGREGATION_TYPE'] == 'School']


grad2015 = grad2015.loc[grad2015['AGGREGATION_TYPE'] == 'District']
grad2016 = grad2016.loc[grad2016['AGGREGATION_TYPE'] == 'District']

```

The function of schoolZip is to give the school district their own zip code. We will perform a join using District name as key. The issue is not all names are a 1 to 1 match. Some names have a slight variance. The frequency of these variances is low enough for me to manually change the name of the original data set to fix these abnormalities. 

```python
#FROM THE IRS I WAS ABLE TO FIND INCOME BY ZIP CODE TO HELP UNDESTAND EACH AREAS INCOME

incomeByZip16 = pd.read_csv('INCOME_DATA/16zpallnoagi.csv')
incomeByZip15 = pd.read_csv('INCOME_DATA/15zpallnoagi.csv')

```

The data I received from the IRS contains not only the data from new york but the data of all of United States. Naturally, I took out the NY rows and replaced the unwanted data from other states.

```python
# df.loc[df['column_name'] == some_value]
incomeByZip16 = incomeByZip16.loc[incomeByZip16['STATE']== "NY"]
incomeByZip15 = incomeByZip15.loc[incomeByZip15['STATE']== "NY"]
```

There are a lot of columns in these data sets. Many of which are not helpful at the time being, especially for my purposes. That being said I took it upon myself to create new data frames but this time with only the data that I foresee myself using. However, if in the future I may require more or less columns, this way I can easily insert them right here. 

```python


cdistrictGrad2016 = pd.DataFrame({
    "District Name" : grad2016["AGGREGATION_NAME"]
})

cdistrictGrad2015 = pd.DataFrame({
    "District Name" : grad2015["AGGREGATION_NAME"]
})
cschoolGrad2016 = pd.DataFrame({
    "School Name" : grad2016Schools["AGGREGATION_NAME"],
    "District Name" : grad2016Schools["LEA_NAME"],
    "ENROLL_CNT": grad2016Schools["ENROLL_CNT"],
    "GRAD_CNT": grad2016Schools["GRAD_CNT"],
    "GRAD_PCT": grad2016Schools["GRAD_PCT"],
    "DROPOUT_PCT" : grad2016Schools["DROPOUT_PCT"],
    "MEMBERSHIP_DESC" : grad2016Schools["MEMBERSHIP_DESC"],
    "SUBGROUP_NAME" : grad2016Schools["SUBGROUP_NAME"]
})
cschoolGrad2015 = pd.DataFrame({
    "School Name" : grad2015Schools["AGGREGATION_NAME"],
    "District Name" : grad2015Schools["LEA_NAME"],
    "ENROLL_CNT": grad2015Schools["ENROLL_CNT"],
    "GRAD_CNT": grad2015Schools["GRAD_CNT"],
    "GRAD_PCT": grad2015Schools["GRAD_PCT"],
    "DROPOUT_PCT" : grad2015Schools["DROPOUT_PCT"],
    "MEMBERSHIP_DESC" : grad2015Schools["MEMBERSHIP_DESC"],
    "SUBGROUP_NAME" : grad2015Schools["SUBGROUP_NAME"]
})

cschoolZip = pd.DataFrame({
    "District Name":schoolZip["District Name"],
    "Zip Code":schoolZip["Zip Code"],
    "Zip +4":schoolZip["Zip +4"]
})

population16 = pd.DataFrame({
    "Zip Code": population["zip_code"],
    "pop": population["y-2016"]
})
population15 = pd.DataFrame({
    "Zip Code": population["zip_code"],
    "pop": population["y-2016"]
})
cschoolGrad2015
```

Here I join the data sets. To join the school zip code DF and graduation DF I used District Name. As for population, I used zip code as my join condition. Also, I joined  the result of the previoilusl 2 joins with my city zipcode (a datasset I created). This is a list of all the zip code within buffalo. The result of this join gave me all the districts in buffalo. Next I remove any duplicates and empty datasets that (usually due to lack of information). Afterwards, I create a list of schools in buffalo and their data using the .isin function and loc. Finally i attach ththe school with the district data.I usually output CSVs in order to look at it in excel. It helps me visualize the data map.

```python
#2016

data = cdistrictGrad2016.join(cschoolZip.set_index('District Name'), on='District Name')
data = data.join(population16.set_index('Zip Code'), on = 'Zip Code')
data = cityZip.join(data.set_index('Zip Code'),on = 'Zip Code')
data = data.dropna(subset=["District Name"])
data = data.drop_duplicates()
cschoolGrad2016 = cschoolGrad2016.loc[cschoolGrad2016['District Name'].isin(data["District Name"])]
data = data.join(cschoolGrad2016.set_index('District Name'), on='District Name')

data.to_csv('MAIN_DATA/2016.csv')



#2015

data2 = cdistrictGrad2015.join(cschoolZip.set_index('District Name'), on='District Name')
data2 = data2.join(population15.set_index('Zip Code'), on = 'Zip Code')
data2 = cityZip.join(data2.set_index('Zip Code'),on = 'Zip Code')
data2 = data2.dropna(subset=["District Name"])
data2 = data2.drop_duplicates()
cschoolGrad2016 = cschoolGrad2015.loc[cschoolGrad2015['District Name'].isin(data2["District Name"])]
data2 = data2.join(cschoolGrad2015.set_index('District Name'), on='District Name')


data2.to_csv('MAIN_DATA/2015.csv')

data2
```

Finally we create all our data sets and save it to the Main Data Folder. We are aggragating the data by cohort groups of 4y 5y and 6y. This will allow us to see if there are any graduation rate discrepencies between each population.

```python
#2016
data2010Cohort = data.loc[data['MEMBERSHIP_DESC']== "2010 Total Cohort - 6 Year Outcome"]
data2011Cohort = data.loc[data['MEMBERSHIP_DESC']== "2011 Total Cohort - 5 Year Outcome"]
data2012Cohort = data.loc[data['MEMBERSHIP_DESC']== "2012 Total Cohort - 4 Year Outcome"]
data2012nAugCohort = data.loc[data['MEMBERSHIP_DESC']== "2012 Total Cohort - 4 Year Outcome - August 2016"]

data2010Cohort.to_csv('MAIN_DATA/2016/Cohort6y.csv')
data2011Cohort.to_csv('MAIN_DATA/2016/Cohort5y.csv')
data2012Cohort.to_csv('MAIN_DATA/2016/Cohort4y.csv')
data2012nAugCohort.to_csv('MAIN_DATA/2016/Cohort4ynAug.csv')


#2015
data2010Cohort = data2.loc[data2['MEMBERSHIP_DESC']== "2009 Total Cohort - 6 Year Outcome"]
data2011Cohort = data2.loc[data2['MEMBERSHIP_DESC']== "2010 Total Cohort - 5 Year Outcome"]
data2012Cohort = data2.loc[data2['MEMBERSHIP_DESC']== "2011 Total Cohort - 4 Year Outcome"]
data2012nAugCohort = data2.loc[data2['MEMBERSHIP_DESC']== "2011 Total Cohort - 4 Year Outcome - August 2015"]

data2010Cohort.to_csv('MAIN_DATA/2015/Cohort6y.csv')
data2011Cohort.to_csv('MAIN_DATA/2015/Cohort5y.csv')
data2012Cohort.to_csv('MAIN_DATA/2015/Cohort4y.csv')
data2012nAugCohort.to_csv('MAIN_DATA/2015/Cohort4ynAug.csv')
```
