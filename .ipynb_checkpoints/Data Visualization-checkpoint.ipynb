# DATA VISUALIZATION

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import scipy.stats as stats
import math
sns.set(style="darkgrid")

# LOADING 2015 Data by Cohort Groups########################################################################

cohort4_2015 = pd.read_csv("MAIN_DATA/2015/Cohort4y.csv")
cohort5_2015 = pd.read_csv("MAIN_DATA/2015/Cohort5y.csv")
cohort6_2015 = pd.read_csv("MAIN_DATA/2015/Cohort6y.csv")
cohort4nA_2015 = pd.read_csv("MAIN_DATA/2015/Cohort4ynAug.csv")


# LOADING 2016 Data Cohort Groups########################################################################

cohort4_2016 = pd.read_csv("MAIN_DATA/2016/Cohort4y.csv")
cohort5_2016 = pd.read_csv("MAIN_DATA/2016/Cohort5y.csv")
cohort6_2016 = pd.read_csv("MAIN_DATA/2016/Cohort6y.csv")
cohort4nA_2016 = pd.read_csv("MAIN_DATA/2016/Cohort4ynAug.csv")


# GET ALL STUDENTS
cohort4_2016_ALL = cohort4_2016.loc[cohort4_2016['SUBGROUP_NAME']== "All Students"]
cohort5_2016_ALL = cohort5_2016.loc[cohort5_2016['SUBGROUP_NAME']== "All Students"]
cohort6_2016_ALL = cohort6_2016.loc[cohort6_2016['SUBGROUP_NAME']== "All Students"]

# 2016 Main Data
main2016 = pd.read_csv("MAIN_DATA/2016.csv")

main2016all = main2016.loc[main2016['SUBGROUP_NAME']=="All Students"]
main2016all['GRAD_PCT'] = main2016all['GRAD_PCT'].str.rstrip('%').astype('float')
main2016all['DROPOUT_PCT'] = main2016all['DROPOUT_PCT'].str.rstrip('%').astype('float')

# 2015 Main Data
main2015 = pd.read_csv("MAIN_DATA/2015.csv")

main2015all = main2015.loc[main2015['SUBGROUP_NAME']=="All Students"]
main2015all['GRAD_PCT'] = main2015all['GRAD_PCT'].str.rstrip('%').astype('float')
main2015all['DROPOUT_PCT'] = main2015all['DROPOUT_PCT'].str.rstrip('%').astype('float')
```

# T TEST

```python
stats.ttest_ind(a=cohort4_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'), b=cohort5_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'),equal_var=False)
```

```python
stats.ttest_ind(a=cohort4_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'), b=cohort6_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'),equal_var=False)
```

```python
stats.ttest_ind(a=cohort5_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'), b=cohort6_2016_ALL["GRAD_PCT"].str.rstrip('%').astype('float'),equal_var=False)
```

With the result of the T-Test it is shown that we cannot reject the notion of the null hypotheis. So we cant say for certain the these two pop;ulation are not the same 
so there are not a significant difference between all the populations
